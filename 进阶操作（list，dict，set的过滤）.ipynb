#### list过滤的方法

```python
data=[-1,2,3,-4,5]
res=[]

for x in data:
    if x>=0:
        res.append(x)
```

```python
print(res)
```

```python
from random import randint
randint(-10,10)
```

```python
data=[randint(-10,10) for _ in range(10)]
print(data)
```

#### 列表解析过滤负数

```python
res=[x for x in data if x>=0]
print(res)
```

#### 过滤器过滤filter

```python
res_=[filter(lambda x:x>=0,data)]
print(res)
```

#### 字典解析

```python
d={'student%d'%i:randint(50,100) for i in range(1,11)}
```

```python
print(d)
```

```python
res={k:v for k,v in d.items() if v>=90}
print(res)
```

#### 字典filter

```python
res_=dict(filter(lambda x:x[1]>=90,d.items()))
print(res_)
```

#### 集合解析

```python
s={randint(0,40) for _ in range(20)}
print(s)
```

```python
res={x for x in s if x%3==0}
print(res)
```
