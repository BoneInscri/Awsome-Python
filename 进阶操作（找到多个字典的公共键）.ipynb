#### 寻找一堆dict中都有的key

```python
from random import randint,sample
p=sample('abcdefg',randint(3,6))
print(p)
```

```python
d1={k:randint(1,3) for k in sample('abcdefg',randint(3,6))}
d2={k:randint(1,3) for k in sample('abcdefg',randint(3,6))}
d3={k:randint(1,3) for k in sample('abcdefg',randint(3,6))}
d4={k:randint(1,3) for k in sample('abcdefg',randint(3,6))}
```

```python
print(d1)
print(d2)
print(d3)
print(d4)
```

```python
for k in d1:
    if k in d2 and k in d3 and k in d4:
        print(k)
```

```python
[k for k in d1 if k in d2 and k in d3 and k in d4]
```

#### 使用map
map(lambda :,?)
先给定一个匹配模式，然后给定匹配的内容

```python
dl=[d1,d2,d3,d4]
print(dl)
[k for k in dl[0] if all(map(lambda d:k in d,dl[1:]))]
```

#### 使用set的交集操作

```python
s1=d1.keys()
s2=d2.keys()
s3=d3.keys()
s4=d4.keys()
print(s1)
print(s2)
print(s3)
print(s4)
```
